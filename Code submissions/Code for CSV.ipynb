```python
import pandas as pd
```

```python
import spacy
```

```python
# Load English model from spaCy
nlp = spacy.load("en_core_web_sm")
```

```python
# Load the list of stop words from the CSV file
stop_words_df = pd.read_csv("C:/Users/fahim/Desktop/stop_words.csv")
```

```python
# Assuming you have imported your CSV file into a DataFrame named 'test'
data = pd . read_csv ( "C:/Users/fahim/Desktop/Test.csv" )
```

```python
# Assuming the stop words are in a column named "Stop_Word"
stop_words_list = stop_words_df["Stop_Words"].tolist()
```

```python
# Add the custom stop words to spaCy's default stop words
for word in stop_words_list:
    nlp.Defaults.stop_words.add(word)
    
```

```python
# Assuming your DataFrame has a column named "comment_processed" containing the text data
texts = data["comment_processed"]
```

```python
# Process each text with spaCy and filter out the stop words
filtered_texts = []
for text in texts:
    doc = nlp(text)
    filtered_text = " ".join(token.text for token in doc if not token.is_stop)
    filtered_texts.append(filtered_text)
```

```python
# Add the filtered text back to the DataFrame
data["filtered_comment"] = filtered_texts
```

```python
# Assuming your DataFrame has a column named "comment_processed" containing the text data
texts = data["comment_processed"]

# Process each text with spaCy and filter out the stop words
filtered_texts = []
for text in texts:
    doc = nlp(text)
    filtered_text = " ".join(token.text for token in doc if not token.is_stop)
    filtered_texts.append(filtered_text)

# Add the filtered text back to the DataFrame
data["filtered_comment"] = filtered_texts
```

```python
# Select 20 rows to compare
rows_to_compare = 20

# Print the original and filtered text side by side for comparison
for index, row in data.head(rows_to_compare).iterrows():
    print("Original Text:")
    print(row["comment_processed"])
    print("\nFiltered Text:")
    print(row["filtered_comment"])
    print("\n---\n")

```
